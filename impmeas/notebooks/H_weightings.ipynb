```python
from matplotlib import pyplot as plt
import numpy as np
xs = np.linspace(0,1,100)
rho1 = 2*np.abs(xs-0.5)
rho2 = 4*(xs-0.5)**2
rho3 = 1+xs*np.log2(xs)+(1-xs)*np.log2(1-xs)
rho3[0] = rho3[-1] = 1
plt.plot(xs, rho1, label="Linear")
plt.plot(xs, rho2, label="Quadratic")
plt.plot(xs, rho3, label="Entropy")
plt.legend()
plt.show()
```
