# Notation

Abzählbar unendliche Menge an Variablen $X = \{ x,y,z,\dots \}$

**Boolsche Vektoren**

* Totale Abbildungen $\mathbf{u},\mathbf{v}, \mathbf{w} : Y \rightarrow \{0,1\}$ für eine endliche Teilmenge $Y \subseteq X$
* Mit Domäne $\text{var}(\mathbf{u})$
* Mit der Länge $| \mathbf{u} | = \sum_x \mathbf{u}(x)$
* Es kann sich auf $Y \subseteq \text{var}(\mathbf{u})$ beschränkt werden: $\mathbf{u}|_Y : Y \rightarrow \{0,1\}$ mit $\mathbf{u}|_Y(x)= \mathbf{u}(x)$ für $x \in Y$
* Konkatenation: für $\text{var}(\mathbf{u}) \cap \text{var}(\mathbf{v}) = \emptyset$ ist $\mathbf{w} = \mathbf{u};\mathbf{v}$ der Vektor mit $\mathbf{w}(x) = \mathbf{u}(x)$ für $x \in \text{var}(\mathbf{u})$ und $\mathbf{w}(x) = \mathbf{v}(x)$ für $x \in \text{var}(\mathbf{v})$
* Für den Fall $\text{var}(\mathbf{u}) = \{ x \}$ schreiben wir auch $\mathbf{u} = (x = c)$ und meinen damit den Vektor mit $\mathbf{u}(x) = c$
* Komposition: für $\text{var}(\mathbf{u}) = \text{var}(\mathbf{v})$ schreiben wir $\mathbf{u} \circ \mathbf{v}$ für $\circ \in \{ \land,\dots\}$ und meinen damit die komponentenweise Verknüpfung $(\mathbf{u} \circ \mathbf{v})(x) = \mathbf{u}(x) \circ \mathbf{v}(x)$.
* Inzidenzvektor: für $Y \subseteq X$ definiert $1_Y : X \rightarrow \{0,1\}$ mit $1_Y(x) = 1$ gdw. $x \in Y$ den Inzidenzvektor für die Menge $Y$; für $Y = \{ y \}$ schreiben wir auch $1_Y = 1_y$.

**Boolsche Funktionen**

* Totale Abbildungen $f, g, h : \{0,1\}^Y \rightarrow \{0,1\}$ für eine endliche Teilmenge $Y \subseteq X$
* Mit der Menge an relevanten Variablen $\text{var}(f) = Y$
* Für $Y \subseteq \text{var}(\mathbf{u})$ werden die Elemente in $\text{var}(\mathbf{u}) \backslash Y$ ignoriert: $f(\mathbf{u}) = f(\mathbf{u}|_Y)$. 
* Komposition $f \circ g$ mit $(f \circ g)(\mathbf{u}) = f(\mathbf{u}) \circ g(\mathbf{u})$ für $\circ \in \{\land, \lor, \dots\}$ und $\text{var}(\mathbf{u}) = \text{var}(f) \cup \text{var}(g)$ 
* Quantifikation: für $\text{var}(f) = Y$ und $x \in Y$ ist $g = \exists x. f$
  * ...die Konstante $g = f(x=0) \lor f(x=1)$ falls $Y = \{ x \}$ oder
  * ...die Funktion $g(\mathbf{u}) = f(\mathbf{u};x=0) \lor f(\mathbf{u};x=1)$ für $\text{var}(\mathbf{u}) = Y \backslash \{ x \}$ falls $\{ x \} \subsetneq Y$
  * kann auf Mengen von Variablen erweitert werden
  * ($\forall$ analog)
* Sei $Y \subseteq X$ eine endliche Menge an Variablen; dann definiert $\mathfrak{F}(Y)$ die Menge an Funktionen mit $\text{var}(f) = Y$.
* Für das Dirac-Delta an der Stelle $\mathbf{u}$ schreiben wir $f_\mathbf{u}$.

# Importance Value

Coalition Games (das sind Tupel $(N, v)$ mit $N$ endl. (i.d.R. auch $N \subseteq \mathbb{N}$) und $n = |N|$ und $v : 2^N \rightarrow \mathbb{R}$ und $v(\emptyset) = 0$) gibt es bereits Importance Values: den Shapley-Wert und den Banzhaf-Wert:

$$ \text{Sh}_x(v) = \sum_{S : x \not\in S} \frac{|S|!(n-|S|-1)!}{n!} (v(S \cup \{ x \}) - v(S)),\quad\quad \text{Bz}_x(v) = \sum_{S : x\not\in S} \frac{1}{2^{n-1}} (v(S \cup \{ x \}) - v(S)) $$

Ein Importance Value für boolsche Funktionen ist eine Abbildung

$$ \gamma : \textstyle\bigcup_{Y \subseteq X \text{ endl.}} \mathfrak{F}(Y) \times Y \rightarrow \mathbb{R} $$

geschrieben $\gamma_x(f)$ für $\gamma(f,x)$.

# Responsibility & Blame

*Responsibility:* Definiert für eine Variable $x$ in Bezug auf einen Vektor $\mathbf{u}$ und eine Funktion $f$ mit $\text{var}(f) = \text{var}(\mathbf{u})$. Wir suchen nach einer Menge $S \subseteq \text{var}(\mathbf{u})$, die den Wert von $f$ unter der Hinzunahme von $x$ ändern kann:

1. $x \not\in S$;
1. Das Kippen der Variablen in $S$ ändert den Wert (noch) nicht: $f( \mathbf{u} \oplus 1_{S} ) = f(\mathbf{u})$;
1. Das zusätzliche Kippen von $x$ aber schon: $f(\mathbf{u} \oplus 1_{S \cup \{ x \}}) \neq f(\mathbf{u})$;
1. Es werden nicht mehr Variablen als notwendig gekippt: es gibt keine kleinere Menge mit diesen Eigenschaften.

Die Responsibility nimmt dann den Wert $\phi^\mathbf{u}_x(f) = 1/(|S|+1)$ an, falls so eine Menge exitiert, und ansonsten 0. Intuitiv gilt

* Falls $|S|$ groß ist, dann sind viele Variablen notwendig, damit $x$ "endlich" Einfluss auf den Wert von $f$ nehmen kann. Damit geht $\phi^\mathbf{u}_x(f) \rightarrow 0$.
* Falls $|S|$ klein ist, dann wird $x$ "schnell wichtig" und $\phi^\mathbf{u}_x(f) \rightarrow 1$.

*Blame:* Sei $Y$ eine Menge an Variablen, $x \in Y$ und $f$ eine Funktion mit $\text{var}(f) = Y$; wir nehmen eine Wahrscheinlichkeitsverteilung $p$ über die Menge der Vektoren über $Y$ und definieren 
$$ \phi_x(f, p) = \textstyle\sum_{\mathbf{u} : Y \rightarrow \{0,1\}} p(\mathbf{u}) \phi^\mathbf{u}_x(f) $$ 

Für eine fixe Verteilung $p$ ist $\phi_x(f,p)$ somit ein Importance Value.


```python
import blame
import binfunc
import tabulate
import ranking

f = binfunc.from_lambda(lambda x,y: x and not y)
print("u=(x, y)")
print(tabulate.tabulate(*blame.responsibilty_table(f)))
print("Blame bei uniformer Verteilung:", ranking.from_val(blame.blame, f))
```


# Coalition Game Mappings

Idee: Importance Values für Coalition Games für unsere Zwecke ausnutzen. Dazu die folgenden "Coalition Game Mappings" definieren, die boolsche Funktionen auf Coalition Games abbilden ($\mathfrak{G}(Y)$ ist die Menge der Coalition Games bei denen $Y$ die Menge der Spieler ist):

$$ \iota: \textstyle\bigcup_{Y \subseteq X,\text{ endl.}} \mathfrak{F}(Y) \rightarrow \textstyle\bigcup_{Y \subseteq X,\text{ endl.}} \mathfrak{G}(Y), $$

mit der Bedingung $\iota(f) = (\text{var}(f), \iota_f)$ (Intuition: eine Variable in $f$ wird zu einem Spieler in $\iota_f$.)

Für passende Übersetzungen bekommen wir Importance Values für boolsche Funktionen "geschenkt": Für ein Coalition Game Mapping $\iota$ wäre bspw. $ \gamma_x(f) = \text{Sh}_x(\iota(f)) $ eine Möglichkeit.

Was sind mögliche Übersetzungen?

**Direktes CGM $\tau_f$**

* Definiert: für $S \subseteq \text{var}(f)$ definiert als $\tau_f(S) = f(1_S)$.
* Nur Funktionen mit $f(0,\dots,0) = 0$ erzeugen gültige Coalition Games.
* Wenn $f$ eine monotone boolsche Funktion ist, dann ist $\tau_f$ ein monotones Coalition Game.
* Alle übersetzten Funktionen ergeben einfache Spiele.
* Ist $f$ self-dual (gilt $f(\mathbf{u}) = \neg f(\neg \mathbf{u})$ für alle $\mathbf{u}$), dann ist $\tau_f$ *proper* und *strong* (d.h. $\tau_f(S) = \neg \tau_f(\overline S)$ gilt für alle $S$).

**Dominierendes CGM $\omega_f$**

* Definiert: für $S \subseteq \text{var}(f)$ definiert als $\omega_f(S) = \exists S. \forall \overline S. f$ (hier ist $\overline S = \text{var}(f) \backslash S$.)
* Es gilt $\omega_f(S) = 1$ wenn die Variablen in $S$ so gewählt werden können, dass $f$ unabhängig von den Variablen in $\overline S$ zu 1 ausgewertet wird.
* Nur falsifizierbare Funktionen ergeben gültige Coalition Games.
* Alle übersetzten Funktionen ergeben monotone und einfache Coalition Games.

**Rektifizierendes CGM $\nu_f$**

* Definiert: für $S \subseteq \text{var}(f)$ definiert als $\nu_f(S) = \forall \overline S. \exists S. f$
* Es gilt $\nu_f(S) = 1$ wenn jede Belegung von Variablen in $\overline S$ durch eine passende Wahl von Variablen in $S$ "korrigiert" werden kann, damit $f$ letztendlich zu 1 ausgewertet wird.
* Nur erfüllbare Funktionen ergeben gültige Coalition Games.
* Alle übersetzten Funktionen ergeben monotone und einfache Coalition Games.

**Symmetrische Dominierendes CGM $\tilde \omega_f$**

* Symmetrische Dominierende Übersetzung: definiert als $\tilde \omega_f(S) = \omega_f(S) \land \omega_{\neg f}(S)$ für $S \subseteq \text{var}(f)$
* Es gilt $\tilde \omega_f(S) = 1$ wenn die Variablen in $S$ so gewählt werden können, dass $f$ immer zu 1 evaluiert, also auch so gewählt werden können, dass $f$ immer zu 0 evaluiert. 
* Jede Funktion ergibt ein gültiges Coalition Game.
* Alle übersetzten Funktionen ergeben monotone und einfache Coalition Games.
* (Rektifizierende analog.)

### Besonderheiten:

Für monotone boolsche Funktionen $f$ gilt $\tau_f = \omega_f = \nu_f$.

* Für $\tau_f(S) = 1$ gilt $f(1_S) = 1$. Damit gibt es aber $\mathbf{s} : S \rightarrow \{0,1\}$, nämlich $\mathbf{s}(x) = 1$ für alle $x \in S$, sodass $f(\mathbf{s};\mathbf{\overline s}) = 1$ für $\mathbf{\overline s}(x) = 0$, $x \in \overline S$. Da $f$ monton ist, gilt aber auch $f(\mathbf{s},\mathbf{\overline s}) \leq f(\mathbf{s},\mathbf{\overline s}')$ für alle $\mathbf{\overline s}' : \overline S \rightarrow \{0,1\}$, und damit $\omega_f(S) = 1$.
* Für $\tau_f(S) = 0$ gilt $f(1_S) = 0$. Angenommen $\omega_f(S) = 1$, es gäbe also $\mathbf{s} : S \rightarrow \{0,1\}$ mit $f(\mathbf{s};\mathbf{\overline s}) = 1$ für alle $\mathbf{\overline s} : \overline S \rightarrow \{0,1\}$. Das gilt dann aber auch insbesondere für $\mathbf{\overline s} = 0$ und damit $1 = f(\mathbf{s}, \mathbf{\overline s}) \leq f(1_S)$. Widerspruch.
* Für $\nu$ kann analog argumentiert werden.

Ist $f$ zusätzlich self-dual, dann ist $\tau_f = \tilde \omega_f = \tilde \nu_f$:

* Von oben wissen wir, dass $\tau_f = \omega_f = \nu_f$ und $\tau_f(S) = \neg \tau_f(\overline S)$ für alle $S$ gilt.
* Für $\tilde \omega_f$ haben wir $\omega_{\neg f}(S) = \exists S. \forall \overline S. \neg f = \neg \forall S. \exists \overline S. f = \neg \nu_f(\overline S) = \neg \tau_f(\overline S) = \tau_f(S) = \omega_f(S).$
* D.h. $\tau_f(S) = \tilde \omega_f(S)$. 
* Analoge Argumentation für $\tilde \nu_f$.

### Importance Values via CGMs

Ein Importance Value für Coalition Games in Kombination mit einem CGM ergibt ein Importance Value für boolsche Funktionen. Abhängig von den obigen Definitionen identifizieren wir einzelne Importance Values über einen dreistelligen Code A/B/C mit A = Sh oder Bz, B = Sym oder ASym, C = Dom oder Rec. Bspw. ist mit Sh/Sym/Dom der Wert $\gamma_x(f) = \text{Sh}_x(\tilde \omega_f)$ gemeint.




```python
import binfunc
import binvec
import ranking
import cgm
import blame

def print_info(f, print_f=True):
  rank1 = ranking.from_val(cgm.get_value("Bz/Sym/Dom"), f)
  rank2 = ranking.from_val(cgm.get_value("Bz/Sym/Rec"), f)
  rank3 = ranking.from_val(blame.blame, f)
  if print_f:
    print(f)
  print("f =", binvec.sum_notation(f.prime_implicants()))
  print("f' =", binvec.sum_notation((~f).prime_implicants()))
  print("-"*100)
  print(rank1, "Bz/Sym/Dom")
  print(rank2, "Bz/Sym/Rec")
  print(rank3, "Blame")

def random_example(eq=True):
  for f in binfunc.iterfunc(3, random=True, maxcount=1000):
    rank1 = ranking.from_val(cgm.get_value("Bz/Sym/Dom"), f)
    rank2 = ranking.from_val(cgm.get_value("Bz/Sym/Rec"), f)
    rank3 = ranking.from_val(blame.blame, f)
    ranks = [rank1, rank2, rank3]
    all_equal = all(rA == rB for rA in ranks for rB in ranks) 
    if eq and all_equal:
      return f
    elif not eq and not all_equal:
      return f

# prints a random example where the rankings are different or equal
# f = random_example(eq=True)

# nice examples
# f = binfunc.from_lambda(lambda x,y,z: x or (y and z) or (not y and not z))
# f = binfunc.from_lambda(lambda x,y,z: x or (y and not z) or (not y and z))
f = binfunc.from_lambda(lambda x,y,z: x or (y and z) )
print_info(f)
```



## Forderungen für "sinnvolle" Importance Values

**Untere/obere Schranken:** Es gilt $0 \leq \gamma_x(f) \leq 1$ für alle $f, x \in \text{var}(f)$.

**Permutationsinvarianz:** Für beliebige $f$, $x \in \text{var}(f)$, Permutationen $\pi : X \rightarrow X$ gilt $\gamma_{\pi(x)}(\pi f) = \gamma_x(f)$. (Dabei ist $\pi f$ die Funktion mit $(\pi f)(\mathbf{u}) = f(\mathbf{u} \circ \pi)$, also $(\mathbf{u} \circ \pi)(x) = \mathbf{u}(\pi(x))$.)

**Dummy:** Für beliebige $f$, $x \in \text{var}(f)$ gilt $\gamma_x(f) = 0$, falls $f(\mathbf{u};x=0) = f(\mathbf{u};x=1)$ für alle $\mathbf{u}$ mit $\text{var}(\mathbf{u}) = \text{var}(f) \backslash \{ x \}$ der Fall ist.

**Diktator:** Für beliebige $f$, $x \in \text{var}(f)$ gilt $\gamma_x(f) = 1$, falls $f(\mathbf{u}) = \mathbf{u}(x)$ oder $f(\mathbf{u}) = \neg \mathbf{u}(x)$ für alle $\mathbf{u}$ mit $\text{var}(\mathbf{u}) = \text{var}(f)$ der Fall ist.

**Negationsinvarianz:** Für beliebige $f$, $x \in \text{var}(f)$ gilt $\gamma_x(\neg f) = \gamma_x(f)$.

**Base Case:** Für alle $\mathbf{u}$ gilt $\gamma_x(f_\mathbf{u}) = \gamma_y(f_\mathbf{u})$ für alle $x,y \in \text{var}(f)$.

### Forderungen für konkrete Reihenfolgen (wann ist eine Variable wichtiger als eine andere)?

**Ranking über Implikanten:** 

Falls $f = f_\mathbf{u} \lor f_\mathbf{v} \lor g$ mit 
* $\text{var}(\mathbf{u}) \cap \text{var}(\mathbf{v}) = \emptyset$ und 
* $\text{var}(g) \cap (\text{var}(\mathbf{u}) \cup \text{var}(\mathbf{v})) = \emptyset$ und 
* $|\text{var}(\mathbf{u})| \leq |\text{var}(\mathbf{v})|$, 
* dann gilt $\gamma_x(f) \geq \gamma_y(f)$ für $x \in \text{dom}(\mathbf{u})$ und $y \in \text{dom}(\mathbf{v})$.  
* Für $|\text{var}(\mathbf{u})| < |\text{var}(\mathbf{v})|$ gilt außerdem $\gamma_x(f) > \gamma_y(f)$.

...das scheint von allen besprochenen Importance Values tatsächlich erfüllt zu werden. Habe aber noch keine fertigen Beweise.
