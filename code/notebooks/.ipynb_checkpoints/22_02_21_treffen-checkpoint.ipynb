## Finale Liste an Eigenschaften

### Definitions

**CHECK THE EXAMPLES**

**Renamings & Complements.** Suppose that $F,H$ are sets of variables of equal size and that $\pi : F \rightarrow H$ a bijection. Call $\pi$ a *renaming*. If $\uf$ is a vector over $F$, define $\pi \uf : x \mapsto \uf(\pi^{-1}(x))$ to be the vector that *translates* $\uf$ to a vector over $H$. Similarly, if $f$ is a function, define $\pi f: \uf \mapsto f(\pi^{-1} \uf)$ to be the function that *translates* $f$ to a function over $H$. 

If $S \subseteq F$ is a set of variables, define $f^S : \uf \mapsto f(\uf \oplus \onef_S)$ to be the function that complements the variables in $S$ before computing $f$.

*Examples.* 
* $f = x \lor y$. Define $\pi(x) = z$, $\pi(y) = y$. Then $\pi f = z \lor y$.
* $f = x \land (y \lor z)$. Define $\pi(x) = y$, $\pi(y) = x$, $\pi(z) = z$. Then $\pi f = y \land (x \lor z)$.
* $f = x \land (y \lor z)$. Define $S = \{ x, y\}$. Then $f^S = \neg x \land (\neg y \lor z)$.

**Types of a Function.** Suppose that $f$ is a function, $\pi : F \rightarrow H$ a renaming and $S \subseteq F$ a set of variables. Call a function $h$ a $(\pi, S)$-equivalent of $f$ if $h = \pi(f^S)$.

*Examples.*
* $f = x \lor y$. Define $\pi(x) = z$, $\pi(y) = y$ and $S = \{ x \}$. Then $h = \neg z \lor y$ is the $(\pi,S)$-equivalent of $f$.
* $f = x \land (y \lor z)$. Define $\pi(x) = y$, $\pi(y) = x$, $\pi(z) = z$ and $S = \{ x \}$. Then $h = \neg y \land (x \lor z)$ is the $(\pi, S)$-equivalent of $f$.

**Extensions.** Call a function $f$ an *extension* of a function $h$ if $H \subseteq F$ and for all $\uf, \uf'$ over $H$: $h(\uf) = h(\uf')$ implies $f[\uf] = f[\uf']$. (Intuition: $f$ only depends on the output of $h$.)

**Monotone Extension.** Call a function $f$ a *monotone extension* of a function $h$ if for all $\uf, \uf'$ over $H$: $h(\uf) \geq h(\uf')$ implies $f[\uf] \geq f[\uf']$. (Intuition: if $f$ evaluates to $0$ even though $h$ evaluates to $1$, there can never be an assignment for the other variables that changes the output of $f$ to $1$.)

*Characterizations of Extensions.*

* $f$ is an extension of $h$ iff there exist functions $s,t$ such that $f = (h \land s) \lor (\neg h \land t)$ and $H \cap (S \cup T) = \emptyset$.
* $f$ is a monotone extension of $h$ iff there exist functions $s,t$ such that $s \geq t$ and $f = (h \land s) \lor (\neg h \land t)$ and $H \cap (T \cup S) = \emptyset$.
* $f$ is a monotone extension of $h$ iff there exist functions $s,t$ such that $f = (h \land s) \lor t$ and $H \cap (T \cup S) = \emptyset$.

*Examples.*
* The function $f = (x \land \neg z_0) \lor (y \land \neg z_0) \lor (\neg x \land \neg y \land z_0 \land z_1)$ is an extension of $h = x \lor y$, as $f = (h \land \neg z_0) \lor (\neg h \land z_0 \land z_1)$.
* It is not a monotone extension however, as $h(x/0; y/1) = 1$ and $h(x/0;y/0) = 0$ and $f(x/0; y/1; z_0/1; z_1/1) = 0$ but $f(x/0; y/0; z_0/1; z_1/1) = 1$.
* The function $f = (x \land y \land z_0) \lor (\neg z_0 \land z_1)$ is a monotone extension of $h = x\land y$.

### Eigenschaften

1. **Type Invariance (TI)**. Let $\gamma$ be an importance value. Then $\gamma_x(f) = \gamma_{\pi(x)}(h)$ for any function $f$ and $(\pi, S)$-equivalent $h$ of $f$.
2. **Complementation Invariance (CI)**. Let $\gamma$ be an importance value. Then $\gamma_x(f) = \gamma_x(\neg f)$.
3. **Decreasing Importance after Monotone Extension (DIME)**. Let $\gamma$ be an importance value. Then $\gamma_x(h) \geq \gamma_x(f)$ for any function $h$ and monotone extension $f$ of $h$.
4. **Rank-Preservation after Monotone Extension (RPME)**. Let $\gamma$ be an importance value. Then $\gamma_x(h) \geq \gamma_y(h)$ implies $\gamma_x(f) \geq \gamma_y(f)$ for any function $h$ and monotone extension $f$ of $h$.
5. **Ranking between Equivalent Modules (REM)**. Let $\gamma$ be an importance value. Suppose that $h$ is a function and $f$ a monotone extension thereof. Let $g$ be a $(\pi,S)$-equivalent of $h$ such that $G \cap F = \emptyset$. Then $\gamma_{\pi(x)}(g * f) \geq \gamma_x(g * f)$ for $* \in \{\land, \lor \}$.
